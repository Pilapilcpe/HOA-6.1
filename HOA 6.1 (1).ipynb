```python
#task 3

def sol(s):
    split_s=list(s[:])
    s_len=len(s)

    if s_len % 2>0:
        split_s.append("_")
    for i in range(len(split_s),0,-1):
        if i % 2:
            split_s[i-1]+= split_s[i]
            split_s.pop(i)
    return split_s
        
print(sol('abc'))
print(sol('abcdef'))
```

```python
#task 3

def inc(s):
    if not s:
        return "1"

    index=len(s) -1
    while index >= 0 and s[index] == " ":
        index-=1

    while index >= 0 and s[index].isdigit():
        index -= 1

    text=s[:index+1]
    number=s[index+1:]

    if number == "":
        return text + "1"

    increment = str(int(number)+1)

    if len(increment) < len(number):
        zeros = len(number) - len(increment)
        increment = "0" * zeros+increment

    return text + increment

print(inc("foo"))
print(inc("foobar"))
print(inc("foo0042"))
print(inc("foo9 "))
print(inc("foo99 "))
    
```
